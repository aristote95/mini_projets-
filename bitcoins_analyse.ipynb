```python
import pandas as pd

file_path = "C:/Users/harmo/OneDrive/Desktop/mini_projet/btcusd_1-min_data.csv"

dtypes = {
    "Open": "float32",
    "High": "float32",
    "Low": "float32",
    "Close": "float32",
    "Volume": "float64"
}

chunksize = 500_000
agg_list = []

for i, chunk in enumerate(pd.read_csv(file_path, dtype=dtypes, chunksize=chunksize)):
    print(f"Chunk {i+1} charg√©, lignes : {len(chunk)}")

    chunk["Timestamp"] = pd.to_datetime(chunk["Timestamp"], unit="s", utc=True)
    
    chunk.set_index("Timestamp", inplace=True)
    chunk.sort_index(inplace=True)

    daily_chunk = chunk.resample("D").agg({
        "Open": "first",
        "High": "max",
        "Low": "min",
        "Close": "last",
        "Volume": "sum"
    })

    agg_list.append(daily_chunk)

btc_d1 = pd.concat(agg_list).resample("D").agg({
    "Open": "first",
    "High": "max",
    "Low": "min",
    "Close": "last",
    "Volume": "sum"
}).dropna()

print(" P√©riode :", btc_d1.index.min(), "‚Üí", btc_d1.index.max())
print(" Jours :", len(btc_d1))
print(btc_d1.head())
print(btc_d1.tail())
print(btc_d1.info())
```
